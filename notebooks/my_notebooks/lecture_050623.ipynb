```python
from io import StringIO
import pandas as pd

table = StringIO("""
T	t	X	a_w	R
-1	-1	 56.11 	0.989	16.93
1	-1	54.89	0.961	50.71
-1	1	28.1	0.942	77.79
1	1	24.81	0.605	151.6
0	0	41.21	0.928	73.83
0	0	43.71	0.928	60.37
0	0	40.38	0.949	74.17
""")
                 
data = pd.read_csv(table, sep='\t')
data
```

```python
from statsmodels.formula.api import ols

#formula = "FPase ~ 1 + C(U, Sum) + C(A,Sum) + C(P,Sum) + C(Y,Sum) + C(U,Sum)*C(A,Sum) + C(U,Sum)*C(P,Sum) + C(U,Sum)*C(Y,Sum) + C(A,Sum)*C(P,Sum) + C(A,Sum)*C(Y,Sum) + C(P,Sum)*C(Y,Sum)"
formula =  "X ~  T+t+T*t"
#formula =  "CMCase ~  1 + C(U) + C(A) + C(P) + C(Y) + C(U)*C(A) + C(U)*C(P) + C(U)*C(Y) + C(A)*C(P) + C(A)*C(Y) + C(P)*C(Y)"
lm = ols(formula, data).fit()

#lm = gls(formula, factor_24_table).fit()
print(lm.summary())
```

```python
from statsmodels.stats.api import anova_lm

anova_lm(lm, typ=3)
```
